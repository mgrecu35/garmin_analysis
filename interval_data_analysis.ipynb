```python
import glob
import pandas as pd
fs=glob.glob('Interval/*')
paceL=[]
strideL=[]
for f in fs:
    df=pd.read_csv(f)
    #print(df)
    pace=df['Avg Pace'].to_numpy()
    stride=df['Avg Stride Length'].to_numpy()
    pace_floatL=[]
    for p in pace:
        p=p.split(':')
        p=float(p[0])+float(p[1])/60
        pace_floatL.append(p)
    paceL.extend(pace_floatL)
    strideL.extend(stride)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.scatter(paceL,strideL)
np.savetxt('paceL.txt',paceL)
np.savetxt('strideL.txt',strideL)
```
